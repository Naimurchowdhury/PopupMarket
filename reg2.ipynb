```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Assuming 'X' is your pandas DataFrame

# Extracting the predictor column
predictor_column = X.iloc[:, -1]

# Extracting the columns to be scaled
columns_to_scale = X.iloc[:, :-1]

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the scaler on the columns to be scaled
scaled_columns = scaler.fit_transform(columns_to_scale)

# Combine the scaled columns with the predictor column
scaled_X = pd.concat([pd.DataFrame(scaled_columns, columns=columns_to_scale.columns), predictor_column], axis=1)

# Creating a correlation matrix
correlation_matrix = scaled_X.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'scaled_X' is your scaled pandas DataFrame

# Determine the optimal number of components using explained variance ratio
pca = PCA()
pca.fit(scaled_X)

# Plotting the explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='-')
plt.title('Explained Variance Ratio')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Determine the optimal number of components
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1

# Perform PCA with the optimal number of components
pca = PCA(n_components=optimal_components)
pca_data = pca.fit_transform(scaled_X)

# Create DataFrame from PCA data
pca_df = pd.DataFrame(data=pca_data, columns=[f'PC{i+1}' for i in range(optimal_components)])

# Create the correlation matrix of the PCA data
pca_correlation_matrix = pca_df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(pca_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of PCA Data')
plt.show()
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'scaled_X' is your pandas DataFrame containing scaled features
# Assuming 'y' is your target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_X.iloc[:,:-1], predictor_column, test_size=0.2, random_state=42)

# Define hyperparameters for grid search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3]
}

# Initialize XGBoost Regressor
xgb_model = XGBRegressor(random_state=42)

# Perform grid search
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Get best hyperparameters
best_params = grid_search.best_params_

# Initialize XGBoost Regressor with best hyperparameters
best_xgb_model = XGBRegressor(random_state=42, **best_params)

# Fit the model to the training data
best_xgb_model.fit(X_train, y_train)

# Predict on the testing data
y_pred = best_xgb_model.predict(X_test)
y_bar = best_xgb_model.predict(X_train)
# Calculate error measures
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)

# Plotting residuals vs fitted values
plt.figure(figsize=(10, 6))
sns.residplot(x=y_bar, y=y_train - y_bar)
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```
